```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.linspace(-20, 30, 100)
yc = x**2 + 7
noise = np.random.randn(x.shape[0]) * 30 
y = yc + noise
plt.plot(x, y, 'o')
```

```python
p = np.polyfit(x, y, deg=2)
pred = np.polyval(p, x)

plt.plot(x, y, 'o')
plt.plot(x, pred, 'r', linewidth=3)
```

```python
#General curve fit
from scipy.optimize import curve_fit
```

```python
x = np.linspace(-20, 30, 100)
yc = 5*x+ 7
noise = np.random.randn(x.shape[0]) * 15
y = yc + noise
plt.plot(x, y, 'o')
```

```python
fun = lambda x, a, b: a*x + b

popt, pcov = curve_fit(fun, x, y)

plt.plot(x, y, 'o')
plt.plot(x, np.polyval(popt, x), 'r', linewidth=3)
```
